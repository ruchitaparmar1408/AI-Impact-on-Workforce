```python
#Setup Notebook first for exploratory data analysis 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your cleaned data
df = pd.read_csv('cleaned_workforce_data.csv')

# Set visual style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
# Calculate average risk per sector
sector_risk = df.groupby('industry_sector')['sector_automation_risk_score'].mean().sort_values(ascending=False)

print(sector_risk)
```

```python
# Plot the bar graph  CHART-1 
plt.figure(figsize=(12,6))
sector_risk.plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('Average Automation Risk Score by Industry Sector')
plt.xlabel('Industry Sector')
plt.ylabel('Average Risk Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Scatter plot — GDP vs Automation Risk Q-2
plt.figure(figsize=(12,6))

plt.scatter(
    df['gdp_per_capita_usd'],
    df['sector_automation_risk_score'],
    alpha=0.5,
    color='steelblue'
)

plt.title('GDP Per Capita vs Automation Risk Score')
plt.xlabel('GDP Per Capita (USD)')
plt.ylabel('Automation Risk Score')
plt.tight_layout()
plt.show()
```

```python
correlation = df['gdp_per_capita_usd'].corr(df['sector_automation_risk_score'])
print(f"Correlation: {correlation:.2f}")
```

```python
print(df['gdp_per_capita_usd'].describe())

# See which countries fall in each cluster
print(df[df['gdp_per_capita_usd'] < 20000]['country'].unique())
print(df[(df['gdp_per_capita_usd'] >= 20000) & (df['gdp_per_capita_usd'] < 60000)]['country'].unique())
print(df[df['gdp_per_capita_usd'] >= 60000]['country'].unique())
```

```python
# Net workforce change by sector
workforce_change = df.groupby('industry_sector')['net_workforce_change_pct'].mean().sort_values()
print(workforce_change)
```

```python


# Color by severity — darker red = worse
plt.figure(figsize=(14,6))
bars = plt.bar(workforce_change.index, workforce_change.values, edgecolor='black')

# Color bars from dark red (worst) to light red (least bad)
colors = plt.cm.Reds_r([i/len(workforce_change) for i in range(len(workforce_change))])
for bar, color in zip(bars, colors):
    bar.set_facecolor(color)

plt.title('Net Workforce Change by Sector\n(All Sectors Showing Negative — AI Destroying More Jobs Than Creating)')
plt.xlabel('Industry Sector')
plt.ylabel('Net Workforce Change %')
plt.axhline(y=0, color='black', linewidth=1.5, linestyle='--')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Find rows where new roles created > displaced
positive_rows = df[df['pct_sector_workforce_new_roles_created'] > df['pct_sector_workforce_displaced']]

print(f"Total rows where creation > displacement: {len(positive_rows)}")
print(positive_rows['industry_sector'].value_counts())
```

```python
# Compare female displacement vs overall displacement by sector
gender_impact = df.groupby('industry_sector')[['pct_sector_workforce_displaced',
                                       'pct_displaced_roles_female']].mean().sort_values('pct_displaced_roles_female', ascending=False)
print(gender_impact)
```

```python
# Visualize side by side
gender_impact.plot(kind='bar', figsize=(14,6), 
                   color=['steelblue','salmon'],
                   edgecolor='black')

plt.title('Overall vs Female Workforce Displacement by Sector')
plt.xlabel('Sector')
plt.ylabel('Displacement Percentage')
plt.legend(['Overall Displaced %', 'Female Displaced %'])
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Compare female workforce % vs female displaced %
female_analysis = df.groupby('industry_sector')[['pct_workforce_female',
                                         'pct_displaced_roles_female',
                                         'pct_sector_workforce_displaced']].mean()

print(female_analysis)
```

```python
adoption_displacement = df.groupby('industry_sector')[['ai_tool_adoption_pct',
                                               'pct_sector_workforce_displaced']].mean()
print(adoption_displacement)
```

```python


plt.figure(figsize=(10,6))
plt.scatter(
    adoption_displacement['ai_tool_adoption_pct'],
    adoption_displacement['pct_sector_workforce_displaced'],
    s=100,
    color='steelblue',
    edgecolor='black'
)

# Add sector labels to each dot
for idx, row in adoption_displacement.iterrows():
    plt.annotate(idx, 
                (row['ai_tool_adoption_pct'], 
                 row['pct_sector_workforce_displaced']),
                fontsize=8)

plt.title('AI Tool Adoption vs Workforce Displacement by Sector')
plt.xlabel('AI Tool Adoption %')
plt.ylabel('Workforce Displaced %')
plt.tight_layout()
plt.show()
```

```python
reskilling_gap = df.groupby('industry_sector')[['sector_automation_risk_score',
                                        'reskilling_programs_count']].mean()
print(reskilling_gap)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(
    reskilling_gap['reskilling_programs_count'],
    reskilling_gap['sector_automation_risk_score'],
    s=100,
    color='red',
    edgecolor='black'
)

for idx, row in reskilling_gap.iterrows():
    plt.annotate(idx,
                (row['reskilling_programs_count'],
                 row['sector_automation_risk_score']),
                fontsize=8)

plt.title('Reskilling Programs vs Automation Risk by Sector')
plt.xlabel('Number of Reskilling Programs')
plt.ylabel('Automation Risk Score')
plt.tight_layout()
plt.show()
```

```python
# Filter India data - INdia specific analysis
india = df[df['iso3_code'] == 'IND']

# India sector risk
india_risk = india.groupby('industry_sector')['sector_automation_risk_score'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
india_risk.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Automation Risk by Sector — India Specific Analysis')
plt.xlabel('Industry Sector')
plt.ylabel('Risk Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
print(india_risk)
```

```python
# Select only numerical columns
numerical_cols = ['gdp_per_capita_usd',
                  'ai_adoption_index', 
                  'sector_automation_risk_score',
                  'pct_sector_workforce_displaced',
                  'net_workforce_change_pct',
                  'ai_skill_wage_premium_pct',
                  'reskilling_programs_count',
                  'govt_ai_policy_score_1_to_10',
                  'ai_tool_adoption_pct',
                  'pct_displaced_roles_female']

# Calculate correlations
correlation_matrix = df[numerical_cols].corr()

# Plot heatmap
plt.figure(figsize=(14,10))
sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt='.2f',
    cmap='RdYlGn',
    center=0,
    linewidths=0.5
)
plt.title('Correlation Matrix — All Key Variables')
plt.tight_layout()
plt.show()
```
